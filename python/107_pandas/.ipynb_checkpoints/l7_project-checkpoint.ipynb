# Data visualization


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

df = pd.read_csv('data/olympics.csv',skiprows=4)
df.head()
```

# In every olympics, which US athelete has won the most total number of medals. Add athlete's discipline

```python
usa=df[df.NOC=='USA']
usa
```

```python
usa.groupby(['Edition','Athlete','Medal']).size()
```

```python
usaMedal = usa.groupby(['Edition','Athlete','Medal']).size().unstack('Medal',fill_value=0)
usaMedal
```

```python
usaMedal['total'] = usaMedal['Gold'] + usaMedal['Silver'] + usaMedal['Bronze']
usaMedal
```

```python
usaMedal.reset_index(inplace=True)
usaMedal
```

```python
for yr, gr in usaMedal.groupby('Edition'):
   #print(gr) 
   print(gr.sort_values('total',ascending=False)[:1]) 
```

```python
mostMedal = [gr.sort_values('total',ascending=False)[:1] for yr, gr in usaMedal.groupby('Edition')]
mostMedal 
```

```python
newDataFrame = pd.DataFrame()
for i in mostMedal:
    newDataFrame=newDataFrame.append(i)
newDataFrame
```
