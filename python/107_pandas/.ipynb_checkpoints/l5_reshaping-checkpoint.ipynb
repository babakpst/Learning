# Reshaping
## Stack: allow you to move the inner columns to rows. 
It returns a DataFrame or Series. Makes the dataframe taller.

## Unstack: reverse of Stack:
Pivot a level of the index labels, returning a DataFrame having a new level of column labels.

If the index is not a MultiIndex, the output will be a Series-the level involved will auto matically get sorted.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('data/olympics.csv',skiprows=4)
df.head()
```

## Athletes winning medals in Beijing Olympics 100m or 200m track

```python
beij= df[(df.Edition == 2008) & ((df.Event == '100m') | (df.Event == '200m')) ]
beij
```

```python
grouped = beij.groupby(['NOC','Gender','Discipline','Event']).size() # output is a series
print(type(grouped),"\n",grouped)
```

```python
mdf = grouped.unstack(['Discipline','Event'])
mdf
```

```python
mdf.stack()
```

```python
mdf.stack('Event')
```

```python
mdf.unstack()
```

```python
mdf.unstack('Gender')
```

# Plot the number of gold medals won by the USA male and female athletes throughout the history of the Olympics

```python
usg=df[(df.NOC=='USA') & (df.Medal=='Gold') ]
usg
```

```python
usg.groupby(['Edition','Gender']).size()
```

```python
usg.groupby(['Edition','Gender']).size().unstack('Gender', fill_value=0)
```

```python
usg.groupby(['Edition','Gender']).size().unstack('Gender', fill_value=0).plot(kind='bar');
```

# plot 5 athletes who won the most medals in the history of olympics. For ties refer to silver then bronze medals.

```python
df.groupby(['Athlete','Medal']).size()
```

```python
medals = df.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
medals
```

```python
medals.sort_values(['Gold','Silver','Bronze'], ascending=False)[['Gold','Silver','Bronze']]
```

```python
medals.sort_values(['Gold','Silver','Bronze'], ascending=False)[['Gold','Silver','Bronze']].head().plot(kind='bar')
```
