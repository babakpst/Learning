```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../../inputs/Environmental_Data_Deep_Moor_2014.csv")
```

Calculate average of a given column for a given month.

```python
Mo_avg_calc = lambda mo,col : df[df['date'].str.contains('201[2345]_[0]?'+ str(mo))][col].mean()
months = range(1,13)
```

```python
wind_speed_2014 = list(map(lambda m: Mo_avg_calc(m, 'Wind_Speed'),months))
```

## Wind Speed Averages 

```python
plt.plot(months,wind_speed_2014)
plt.show()
```

```python
wind_speed_2014 = list(map(lambda m: Mo_avg_calc(m, 'Wind_Speed'),months))
wind_gust_2014 = list(map(lambda m: Mo_avg_calc(m,'Wind_Gust'),months))
dew_point_2014 = list(map(lambda m: Mo_avg_calc(m, 'Dew_Point'),months))
pressure_2014 = list(map(lambda m: Mo_avg_calc(m, 'Barometric_Press'),months))
```

```python
plt.plot(wind_speed_2014,label='Wind Speed',color="#2BD31444")
plt.plot(wind_gust_2014,'d-',label='Wind Gust',color="#2BD314FF")
plt.plot(dew_point_2014,'d-',label='Dew Point',color='#142BD3FF')
plt.plot(pressure_2014,',:',label='Barometric Pressure',color="#142BD366")
plt.ylim(0,80)
plt.legend()
plt.show()
```

## Alternate Styling Options

```python
plt.plot(wind_speed_2014,label='Wind Speed',color="#2BD31444")
plt.plot(wind_gust_2014,'d-',label='Wind Gust',color="#2BD314FF")
plt.plot(dew_point_2014,'d-',label='Dew Point',color='#142BD3FF')
plt.plot(pressure_2014,',:',label='Barometric Pressure',color="#142BD366")
plt.ylim(0,80)
#plt.legend()
plt.title('Yearly trends')
plt.axis('off')
plt.text(8,60, 'Dew Point')
plt.text(6,20, 'Wind Gust')
plt.show()
```
