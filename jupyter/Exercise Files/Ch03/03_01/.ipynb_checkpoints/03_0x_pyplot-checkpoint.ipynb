```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv("../../inputs/Environmental_Data_Deep_Moor_2015.csv")
```

Wrangle data into arrays for both x and y axis

```python
feb15 = df[df['date']=='2015_02_15']
feb15_hours = [time.split(':')[0] for time in feb15['time']]
feb15_hours_dict = {i:feb15_hours.count(str(i)) for i in np.arange(24)}
```

```python
plt.plot(feb15_hours_dict.keys(),feb15_hours_dict.values())
```

```python
feb16 = df[df['date']=='2015_02_17']
feb16_hours = [time.split(':')[0] for time in feb16['time']]
feb16_hours_dict = {i:feb16_hours.count(str(i)) for i in np.arange(24)}
```

```python
plt.plot(feb15_hours_dict.keys(),feb15_hours_dict.values(),feb15_hours_dict.keys(),feb16_hours_dict.values())
plt.xticks(np.arange(0,24,6))
plt.xlim(-3,26)
plt.ylim(-3,28)
plt.xlabel('Hour of the Day')
plt.ylabel('Measurements Taken')
plt.title('Measurements by Hour')
plt.show()
```
